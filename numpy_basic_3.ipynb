```python
w = np.array([[1,2,3,4],[99,34,67,56],[21,13,18,22]])
print(w)
```

```python
print(w[w<5])
```

```python
pc = (w>=5)
print(w[pc])
```

```python
pv = (w%2==0)
print(w[pv])
```

```python
cw = w[(w>2) & (w<21)]
print(cw)
```

```python
sw = w[(w<8) | (w>50)]
print(sw)
```
